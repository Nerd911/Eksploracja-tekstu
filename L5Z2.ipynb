```python
def to_lemma_mapping():
    all_lemmas = dd(list)
    for line in open('Dane/polimorfologik-2.1.txt', encoding='utf-8'):
        L = line.split(';')[:2]
        all_lemmas[L[1].lower()].append(L[0].lower())
    return all_lemmas
```

```python
def preprocess_article(wiki_list, stopwords):
    overall_counts = dd(lambda:0)
    document_frequency = dd(lambda:0)
    article_counts = []
    for title, article in wiki_list:
        article_count = dd(lambda:0)
        for line in article:
            for w in line:
                w = w.lower()
                if not w.isalpha():
                    continue
                if w in stopwords:
                    continue
                if w not in lemmas:
                    lemmas[w] = [w]
                for el in lemmas[w]:
                    overall_counts[el] += 1
                    if el not in article_count:
                        document_frequency[el] += 1
                    article_count[el] += 1
        article_counts.append(article_count)
    N = len(overall_counts)
    indices = {}
    M = len(article_counts)
    of_the_jedi = lil_matrix((M, N))
    
    for i in range(M):
        for j, k in enumerate(overall_counts.keys()):
            of_the_jedi[i, j] = (article_counts[i][k]/overall_counts[k]) * np.log(N/(document_frequency[k] + 1))
    return of_the_jedi
```

```python
processed = preprocess_article(wiki_list, stopwords)
```
