```python
pattern = re.compile(r"''[a-zA-Z0-9_ ]+''")
candidates = dd(list)
for i, line in enumerate(open('data/poczatki_wikipediowe.txt', 'r', encoding='utf-8')):
    if line == '\n':
        continue
        
    if line[:3] == '###':
        line = line.split(' ')[1:]
        if len(line) > 1:
            header = None
            continue
        header = line[0][:-1]
        
    elif header is not None:
        words = pattern.findall(line)
        
        if len(words) == 0:
            continue
        result = []
        for w in words:
            if header != w[2:-2] and len(w[2:-2].split(' ')) == 1:
                result.append(w[2:-2])
        if len(result) > 1:
            candidates[header] = result
```

```python
def get_common_letters_number(s1, s2):
    count = 0
    for ch in s1:
        if ch in s2:
            count += 1
    return count
```

```python
for key in tqdm_notebook(candidates):
    key_emb = word2vec[key.lower()]
    if key_emb == []:
        continue
    
    result = []
    for word in candidates[key]:
#         if word.lower()[:3] == key.lower()[:3]:
#             continue
        if get_common_letters_number(word.lower(), key.lower()) >= len(word) * 0.75:
            continue
        
        if len(word) <= 2:
            continue
            
        word_emb = word2vec[word.lower()]
        if word_emb == []:
            continue
        distance = linear_kernel(np.array(word_emb).reshape(1,-1), np.array(key_emb).reshape(1,-1))
        
        if distance[0][0] > 30:
            result.append(word)
            
    if len(result) == 0:
        continue
        
    print(' # '.join([key] + list(np.unique(np.array(result)))))
```
